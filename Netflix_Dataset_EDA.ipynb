<h2>Data Cleaning & Preprocessing</h2>

The dataset was cleaned by handling missing values, standardizing column formats, and converting duration fields into numeric values. Missing categorical values such as country, director, and cast were imputed using appropriate techniques like mode imputation or labeled as “Unknown” where necessary. Duplicate records were checked, and inconsistent data types were corrected to ensure data quality and accuracy for analysis.

```python
missing_values=df1[df1['director'].isna()| df1['cast'].isna()| df1['country'].isna()]
print('Number of Titles having missing director,caste,country are',len(missing_values))
```

```python
df1['director']=df1['director'].fillna('unknown')
df1['director'].isna().sum()
```

```python
df1['country']=df1['country'].fillna('unknown')
df1['country'].isna().sum()
```

```python
df1['cast']=df1['cast'].fillna('unknown')
df1['cast'].isna().sum()
```

Missing values in the director,cast,country columns were imputed with “Unknown” to preserve records while clearly indicating unavailable information.

```python
df1['rating']=df1['rating'].dropna(axis=0)
```

```python
df1['duration']=df1['duration'].dropna(axis=0)
```

```python
df1=df1.dropna(subset=['date_added'])
```

```python
df1=df1.copy()
df1['date_added']=pd.to_datetime(df1['date_added'],format='mixed')
df1['date_added']
```

```python
df1['date_added'].isna().sum()
```

```python
df1['year_added']=df1['date_added'].dt.year
```

```python
df1['month_added']=df1['date_added'].dt.month_name()
```

```python
df1['duration']=df1['duration'].astype('str').str.extract(r'(\d+)').astype('float')
df1['duration']
```

```python
country_split=df1['country'].str.split(',').explode().str.strip().reset_index(drop=True)
country_split
```

```python
listed_in_split=df1['listed_in'].str.split(', ').explode().str.strip().reset_index(drop=True)
listed_in_split
```

<h2>Exploratory Data Analysis (EDA)</h2>

Exploratory Data Analysis was conducted using univariate and bivariate analysis techniques. Distribution of content types, ratings, release years, duration, and genres were analyzed using bar plots, count plots, histograms, and line charts. Comparisons between movies and TV shows were performed to identify dominant content types and rating patterns on Netflix.

```python
#Univariate Analysis
plt.figure(figsize=(3,4))
plt.hist(df1['duration'],color='r')
plt.xlabel('Duration')
plt.ylabel('Count_of_Shows')
plt.show()
```


The genres with the highest number of titles on Netflix are Dramas, International Movies, and Documentaries, each accounting for approximately 350 entries. This suggests that these genres are the primary contributors to Netflix’s content offerings. For TV shows Kid's TV is the most common genre.

```python
most_frequent=df1['cast'].value_counts().reset_index().head(5)
plot_data=most_frequent[most_frequent['cast']!='unknown']
plot_data
```

David Attenborough is the most frequently featured cast member in the Netflix dataset, reflecting Netflix’s strong focus on documentaries.

```python
df_country=df1.assign(country=country_split)
df_country['country']=df_country['country'].str.lower()
country_wise_type=df_country.groupby(['type','country']).size().reset_index(name='Count')
plot_chart=country_wise_type[country_wise_type['country']!='unknown']
top10_countries=plot_chart.groupby('country')['Count'].sum().sort_values(ascending=False).head(10).reset_index()
top_chart=plot_chart[plot_chart['country'].isin(top10_countries['country'])]
plt.figure(figsize=(8,6))
sns.barplot(x='country',y='Count',data=top_chart,hue='type')
plt.xticks(rotation=45)
plt.title('Top_10_countries_by_Netflix_Content')
plt.tight_layout()
plt.show()
```

Analysis of country-wise content shows that the United States produces the most Netflix content, particularly movies. India follows as the second-largest contributor.

```python
df1['age_partition']=df1['year_added']-df1['release_year']
sns.histplot(df1['age_partition'].dropna(),bins=30)
```

We compared the release year of titles with the year they were added to Netflix. Most titles are added after their release, with some content being older classics. This indicates that Netflix acquires both recent and older titles for its library.

```python
after_2010=df1[df1['year_added']>2010]
before_2010=df1[df1['year_added']<=2010]
counts=[len(after_2010),len(before_2010)]
labels=['After_2010','Before_2010']
plt.bar(labels,counts,color=['orange','red'])
plt.ylabel('Number of Titles')
plt.title('Titles Added to Netflix: Before vs After 2010')
plt.show()
```

The dataset shows that nearly all titles were added to Netflix after 2010, while very few titles were added before 2010. This reflects the platform’s growth in content acquisition during the last decade.

```python
after_2015=df1[df1['year_added']>2015]
df_country=after_2015.assign(country=country_split)
df_country['country']=df_country['country'].str.lower()
country_wise_type=df_country.groupby(['type','country']).size().reset_index(name='Count')
plot_chart=country_wise_type[country_wise_type['country']!='unknown']
top5_countries=plot_chart.groupby('country')['Count'].sum().sort_values(ascending=False).head(5).reset_index()
top_chart=plot_chart[plot_chart['country'].isin(top5_countries['country'])]
plt.figure(figsize=(8,6))
sns.barplot(x='country',y='Count',data=top_chart,hue='type')
plt.xticks(rotation=45)
plt.title('Top 5 Countries by Netflix Content (After 2015)')
plt.ylabel('Number of Titles')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

After 2015, the top 5 countries producing Netflix content are the United States, India, the United Kingdom, Canada, and France. Movies dominate in most countries, while TV Shows contribute significantly in a few. This reflects Netflix’s expansion strategy in different regions during recent years.

```python
df_genres=df1.copy()
df_genres['genre']=listed_in_split
df_genres['year']=df1['year_added']
genre_trend=df_genres.groupby(['genre','year']).size().reset_index(name='Count')
top5_genres=df_genres['genre'].value_counts().head(5).reset_index()
top_trend=genre_trend[genre_trend['genre'].isin(top5_genres['genre'])]
plt.figure(figsize=(12,6))
sns.lineplot(data=top_trend, x='year', y='Count', hue='genre', marker='o')
plt.title('Top 5 Netflix Genres Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

Most of the content additions on Netflix occurred between 2018 and 2020, with a significant increase in international movies during this period. Afterward, Drama became the most preferred genre, which is why Drama content ranks second overall.

```python
df_rating=df1.copy()
df_rating['rating']=df1['rating']
df_rating['year']=df1['year_added']
rating_trend=df_rating.groupby(['rating','year']).size().reset_index(name='Count')
top5_rating=df_rating['rating'].value_counts().head(5).reset_index()
top_trend=rating_trend[rating_trend['rating'].isin(top5_rating['rating'])]
plt.figure(figsize=(12,6))
sns.lineplot(data=top_trend, x='year', y='Count', hue='rating', marker='o')
plt.title('Top 5 Netflix Ratings Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.legend(title='Ratings')
plt.tight_layout()
plt.show()
```

Netflix content is mostly aimed at mature audiences. TV-MA titles are the most frequent, totaling around 850, while TV-14 titles come second with about 550. This suggests that a significant portion of the library targets adult viewers.

```python
without_null=df1[df1['country']!='unknown']
top5_countries=without_null['country'].value_counts().head().reset_index()
top_country_data=without_null[without_null['country'].isin(top5_countries['country'])]
country_rating_counts = top_country_data.groupby(['country','rating']).size().reset_index(name='Count')
plt.figure(figsize=(10,6))
sns.barplot(x='country', y='Count', hue='rating', data=country_rating_counts)
plt.xticks(rotation=45)
plt.title('Content Rating Distribution Across Top 5 Countries')
plt.ylabel('Number of Titles')
plt.xlabel('Country')
plt.legend(title='Rating')
plt.tight_layout()
plt.show()
```

As illustrated in the chart above, Netflix content ratings vary across countries. The United States mostly has TV-MA and R titles, while India favors TV-14, with TV-MA coming second. This highlights differences in content preference and regulatory environments between countries.

<h2>Key Insights from the Analysis</h2>

The analysis revealed that Movies constitute the majority of Netflix content compared to TV Shows. TV-MA and TV-14 are the most common ratings, indicating a strong focus on mature and teen audiences. Most movies have a duration between 50 and 150 minutes. Content addition peaked between 2015 and 2020, after which a decline is observed. The United States contributes the highest number of titles, followed by other major content-producing countries.

<h2>Business Recommendations / Conclusion</h2>

Based on the analysis, Netflix should continue investing in TV-MA and TV-14 rated content, as these categories dominate the platform. Increasing TV show production could help improve long-term user engagement. Netflix can also explore content expansion in underrepresented countries to diversify its global catalog. Understanding duration and rating preferences can further help optimize content acquisition strategies.
